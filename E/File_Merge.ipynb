# 파일 합치기

```python
import os 
import pandas as pd

import warnings
warnings.filterwarnings(action='ignore')
```

폴더 내에 있는 파일 하나로 합치기

```python
kwd = pd.DataFrame(columns = ['Keyword', 'SubPatternName'])

# 폴더 내에 있는 파일 목록 불러오기
for f in os.listdir('.'):
    # .txt 확장자만
    if f[-4:] == '.txt':
        temp = pd.read_csv(f, names = ['Keyword'])
        # 파일명에서 추출
        temp['Pattern'] = f.replace('_0928.txt', '')
        # 하나의 파일로 합치기
        kwd = pd.concat([kwd, temp])
```

Group 번호 붙이기

```python
# 파일 불러오기
group = pd.read_excel('Group.xlsx')

# Group 번호 붙이기
kwd = pd.merge(kwd, group[['Group', 'SubPatternName']], on = 'SubPatternName', how = 'left')

# 정렬
kwd = kwd[['Group', 'SubPatternName', 'Keyword']].sort_values(['Group', 'Keyword']).reset_index(drop = True)
```

내보내기

```python
kwd.to_excel('###파일명###.xlsx', index = False)
```
