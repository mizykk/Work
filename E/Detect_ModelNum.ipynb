# 모델명 찾기

```python
import re
import pandas as pd

import warnings
warnings.filterwarnings(action = 'ignore')
```

데이터 불러오기

```python
data = pd.read_excel('파일명.xlsx', header = 1)
```

작업해야하는 데이터만 남기기

```python
data = data[data['구분']=='작업대상'].reset_index(drop = True)
data.sample(3)
```

숫자로만 이루어진 데이터 → 제외

```python
# 데이터타입이 int면 '숫자' 출력

data['type'] = data['keyword'].apply(lambda x: '문자' if type(x) != int else '숫자')
```

```python
# 숫자 데이터 제외
data = data[data['type']!='숫자']
```

### 영여 & 숫자 & - 만 허용

```python
com = re.compile('^([a-zA-Z-_])+([0-9-_])+([a-zA-Z0-9])*$')

data['모델명추정'] = data['keyword'].apply(lambda x: com.match(x).group() if com.match(x) != None else '')
```

내보내기

```python
data[data['모델명추정']!=''].sample(3)
```

```python
data[data['모델명추정']!=''].to_excel('모델명.xlsx', index = False)
```
